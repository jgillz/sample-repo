## Dictionaries Lab Solutions

**Step 1:  Fetch Data Off Internet**

```python
import requests

url = 'https://api.census.gov/data.json'

# you can use the requests get() method like this to pull in json data
data = requests.get(url).json()
```

```python
# and this is what it will look like
data
```

**Step 2:  Locate the Actual Dataset**

```python
# it's here -- notice the repeating dictionaries inside the list
data['dataset']
```

**Step 3: Load It Into a DataFrame**

```python
import pandas as pd

# pass it into the dataframe method like so
df = pd.DataFrame(data['dataset'])
```

```python
# and check its head like this
df.head()
```

**Step 4:  Cleaning It Out**

```python
bureau_Codes = []

for item in data['dataset']:
    try: 
        if item['bureauCode'][0]:
            bureau_Codes.append(item['bureauCode'][0])
    except:
        bureau_Codes.append(None)
```

**Step 5:  Creating a New Column**

```python
# this sets the value of this column to this list
df['bureauCode'] = bureau_Codes
```

```python
# and we can see that our column is now 'clean'
df.head()
```
