```python
import pandas as pd
import numpy as np

superstore = '../data/superstore.xls'

df = pd.read_excel(superstore)

df.dtypes
```

```python
today = np.datetime64('today')
#column 1
df2 = df.groupby('Customer ID')['Sales','Profit'].sum() 

#column 2, need help - need first purchase, groupby customer id and find sort by earliest order date
time_frame = (df.groupby('Customer ID')['Order Date'].min()).to_frame()
df2 = df2.merge(time_frame, on='Customer ID', how='left')
df2['Time As Customer'] = today - df2['Order Date']

#coulmn 3
df2['Total Order Count'] = df.groupby('Customer ID')['Sales'].count()

#column 4 needs col 2

#column 5
df2['Repeat Customer'] = df2['Total Order Count'] > 1

#column 6
df['Segment'].unique()
df['Consumer'] = df['Segment'] == 'Consumer'
df['Home Office'] = df['Segment'] == 'Home Office'
df['Corporate'] = df['Segment'] == 'Corporate'
cust_type = df.groupby('Customer ID')['Consumer','Home Office','Corporate'].sum().idxmax(axis=1).to_frame() #is a way to find total sales per type for each customer
df2 = df2.merge(cust_type, on='Customer ID', how='left')

#column 7 need col 2
#col 8 needs col 2
#col 9 needs col 2
#col 10 needs col 2
#col 11 needs col 2

df2
```
