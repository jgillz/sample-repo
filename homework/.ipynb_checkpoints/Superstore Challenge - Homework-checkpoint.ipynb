```python
import pandas as pd
import numpy as np
import datetime

superstore = '../data/superstore.xls'

df = pd.read_excel(superstore)

df.dtypes
```

```python
today = np.datetime64('today')
#column 1
df2 = df.groupby('Customer ID')['Sales','Profit'].sum() 

#column 2, need help - need first purchase, groupby customer id and find sort by earliest order date
time_frame = (df.groupby('Customer ID')['Order Date'].min()).to_frame()
df2 = df2.merge(time_frame, on='Customer ID', how='left')
df2['Time As Customer'] = today - df2['Order Date']

#coulmn 3
df2['Total Order Count'] = df.groupby('Customer ID')['Sales'].count()

#column 4 create cohor
df2['Cohort'] = np.where(pd.DatetimeIndex(df2['Order Date']).month >=7,2,1)

#column 5
df2['Repeat Customer'] = df2['Total Order Count'] > 1

#column 6
df['Segment'].unique()
df['Consumer'] = df['Segment'] == 'Consumer'
df['Home Office'] = df['Segment'] == 'Home Office'
df['Corporate'] = df['Segment'] == 'Corporate'
cust_type = df.groupby('Customer ID')['Consumer','Home Office','Corporate'].sum().idxmax(axis=1).to_frame() #is a way to find total sales per type for each customer
df2 = df2.merge(cust_type, on='Customer ID', how='left')
df2['Customer ID2'] = df2.index
#column 7 3 months later
df2['3 Months Later'] = df2['Order Date'] + np.timedelta64(3,'M')
#col 8 needs col 2
df2['1 Year Later'] = df2['Order Date'] + np.timedelta64(1,'Y')

#how do i avoid re-making the columnms below?
year_later = df2['1 Year Later'].to_frame()
three_months_later = df2['3 Months Later'].to_frame()
df = df.merge(three_months_later, on='Customer ID',how='left')
df = df.merge(year_later, on='Customer ID', how = 'left')

#col 9 - make a column in df that shows 90 days after the first purchase, if the count of purchases is > 1 for all purchases within 90 days, show those customers
df['purchase within 90 days'] = df['Order Date'] < df['3 Months Later'] # this isnt correct
#col 10 - remove all dates greater than 1 year, count purchases inside that

#col 11 filter for all dates after 1 year, sum sales

```

```python
df2['First Order'] = df2['Order Date']
first_order = df2['First Order'].to_frame()
first_order
```

```python
df = df.merge(first_order,on='Customer ID',how='left')
excluding_first = df[~(df['Order Date'] == df['First Order'])]
```

```python
made_purchase_within_90d = excluding_first.groupby('Customer ID')['purchase within 90 days'].sum() >0
made_purchase_within_90d = made_purchase_within_90d.to_frame()
```

```python
df2 = df2.merge(made_purchase_within_90d, on='Customer ID',how='left') #make column 9
```

```python
orders_in_first_year = df[df['Order Date'] < df['1 Year Later']].groupby('Customer ID')['Order ID'].count().to_frame()
df2 = df2.merge(orders_in_first_year, on= 'Customer ID',how='left') #make column 10
```

```python
sum_of_first_year = df[df['Order Date'] < df['1 Year Later']].groupby('Customer ID')['Sales'].sum().to_frame()
df2 = df2.merge(sum_of_first_year,on='Customer ID',how='left') #column 11, named sales_y
```
